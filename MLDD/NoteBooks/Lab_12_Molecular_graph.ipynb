# Molecules to Graph


In this section we will be using the rdkit mol object to generate the following features for our property predcition task

> Node_Features

> Edge_list

> Edge_Features

We use some utility functions described below for genrating the input for our model which is a tuple of the above three

```python
def featurize_atom(atom: rdkit.Chem.Atom) -> np.ndarray:
    """
    Generates a Feature Vector for the given atom
    :param atom: atom object in a rdkit molecule
    :return: np.array, the feature vector for the atom
    """
    possible_atom_labels = np.array(
        ["C", "N", "O", "F"]
        # list_of_atoms
        
    )
    atom_label = atom.GetSymbol() == possible_atom_labels

    feature_vector = np.concatenate([atom_label]).astype(np.float64)
    return feature_vector
```

```python
def featurize_bond(bond: rdkit.Chem.Bond) -> np.ndarray:
    """
    Generates a Feature Vector for the given bond
    :param bond: bond object in a rdkit molecule
    :return: np.array, the feature vector for the atom
    """
    possible_bond_types = np.array(
        [
            rdkit.Chem.rdchem.BondType.SINGLE,
            rdkit.Chem.rdchem.BondType.DOUBLE,
            rdkit.Chem.rdchem.BondType.TRIPLE,
            rdkit.Chem.rdchem.BondType.AROMATIC,
        ]
    )
    bond_type = bond.GetBondType() == possible_bond_types

    feature_vector = np.concatenate([bond_type]).astype(np.float64)
    return feature_vector
```

```python
def featurize_molecule(molecule, permute=True):
    """
    Method that constructs a molecular graph with nodes being the atoms
    and bonds being the edges.
    :param molecule: RD-kit molecule object to featurize
    :param permute: permute the atoms if True, otherwise leave in order
    :return: PyG graph object, Node features and Edge features
    """
    num_atoms = molecule.GetNumAtoms()
    permute = np.random.permutation(num_atoms) if permute else np.arange(num_atoms)

    node_features, edge_features, edge_list = [], [], []
    for i in range(molecule.GetNumAtoms()):
        atom_i = molecule.GetAtomWithIdx(int(permute[i]))
        node_features.append(featurize_atom(atom_i))
        for j in range(molecule.GetNumAtoms()):
            bond_ij = molecule.GetBondBetweenAtoms(int(permute[i]), int(permute[j]))
            if bond_ij is not None:
                edge_list.append([permute[i], permute[j]])
                edge_features.append(featurize_bond(bond_ij))

    if len(edge_list) == 0:
        return (
            torch.from_numpy(np.stack(node_features)).float(),
        )

    return (
        torch.from_numpy(np.stack(node_features)).float(),
        torch.from_numpy(np.stack(edge_list, axis=1)).long(),
        torch.from_numpy(np.stack(edge_features)).float(),
    )
```

# Graph Neural Networks from Scratch

Here Graph is a data structure which contains two components Vertices/Nodes and edges(directed or undirected)

> G = (V, E)

Graph Neural Network is a type of Neural Network which directly operates on the Graph structure. 


## GNN base class

It's always a good idea to write down the basics of how a message passing neural network (MPNN) works, before we dive deep into the specifics of a particular architecture. 

Here we will take 4 tasks that an MPNN does as abstraction.

```python
class GraphNeuralNetwork(torch.nn.Module, abc.ABC):
    def __init__(
        self,
        in_channels,
        hidden_channels,
        out_channels,
        layers,
        adj_normalizer="row"
    ):
        super().__init__()

        assert adj_normalizer in ["row", "column", "symmetric", "none"]
        self.adj_normalizer = adj_normalizer

        self.in_channels = in_channels
        self.hidden_channels = hidden_channels
        self.out_channels = out_channels
        self.layers = layers

    def forward(self, x, edge_index):
        # Get the input node features into a node feature vector of hidden-size 
        x = self.initialize(x)

        # Generate the degree for normalization
        degrees = np.zeros(shape=(len(x),))
        for edge in edge_index.T:
            degrees[edge[0]] += 1

        # Apply the graph neural network layers
        for layer_idx in range(self.layers):
            # Aggregate features from all neighbors
            aggregated_neighbors = [[] for _ in range(x.shape[0])]
            for edge in edge_index.T:
                aggregated_neighbors[edge[0]].append(
                    x[edge[1]]
                    / self.get_degree_normalizer(degrees[edge[0]],
                                                 degrees[edge[1]])
                )
            # Apply neural conv transform and pool across those neighbors
            final_outputs = []
            for i in range(len(x)):
                aggregated_neighbor_tensor = self.aggregate(
                    aggregated_neighbors[i], layer_idx
                )
                final_outputs.append(
                    self.combine(
                        aggregated_neighbor_tensor,
                        x[i],
                        layer_idx,
                    )
                )
            # Recover the feature matrix from the new node features
            x = torch.stack(final_outputs, dim=0)

        # Pool the output across all the nodes and get the final output
        x = self.output(x)

        # That's the result
        return x

    def get_degree_normalizer(self, degree_u, degree_v):
        degree_u += 1
        degree_v += 1
        if self.adj_normalizer == "row":
            return degree_u
        elif self.adj_normalizer == "column":
            return degree_v
        elif self.adj_normalizer == "symmetric":
            return (degree_v * degree_u) ** 0.5
        elif self.adj_normalizer == "none":
            return 1
        else:
            raise ValueError("Invalid Adjacency Normalizer")

    @abc.abstractmethod
    def initialize(self, node_feature_tensor):
        raise NotImplementedError

    @abc.abstractmethod
    def aggregate(self, aggregated_neighbors, layer_idx):
        raise NotImplementedError

    @abc.abstractmethod
    def combine(self, aggregated_neighbor_tensor, self_tensor, layer_idx):
        raise NotImplementedError

    @abc.abstractmethod
    def output(self, node_feature_tensor):
        raise NotImplementedError
```

```python
class GCN(torch.nn.Module):
    def __init__(self, in_feats, hidden_size, num_classes, mode):
        super(GCN, self).__init__()
        self.conv1 = dgl.nn.pytorch.GraphConv(in_feats, hidden_size)
        self.conv2 = dgl.nn.pytorch.GraphConv(hidden_size, hidden_size)
        self.conv3 = dgl.nn.pytorch.GraphConv(hidden_size, num_classes)

        if mode == "binary":
            self.output_activation = torch.nn.Sigmoid()
        elif mode == "classification":
            self.output_activation = torch.nn.Softmax(dim=1)
        elif mode == "regression":
            self.output_activation = torch.nn.Identity()

    def forward(self, g, inputs):
        h = self.conv1(g, inputs)
        h = torch.relu(h)
        h = self.conv2(g, h)
        h = torch.relu(h)
        h = self.conv3(g, h)
        h = self.output_activation(h)
        h = torch.mean(h, dim=0)
        return h
```

```python
def train_dgl(model, dataset, loss_fn):
    optimizer = torch.optim.Adam(model.parameters())
    train_dataset = dataset[:int(0.8*len(dataset))]
    validation_dataset = dataset[int(0.8*len(dataset)):]

    train_losses, val_losses = [], []
    for epoch in range(10):
        # Training step
        model.train()
        train_loss, train_items = 0.0, 0
        val_loss, val_items = 0.0, 0
        with tqdm.tqdm(train_dataset,
                       desc=f'Training Epoch {epoch}') as train_iterator:
            for mol, y in train_iterator:
                x = featurize_molecule(rdkit.Chem.MolFromSmiles(mol))
                try:
                  g = dgl.graph((x[1][0], x[1][1]))
                  g.ndata['x'] = x[0]
                  y = torch.tensor([y])
                  optimizer.zero_grad()
                  prediction = model(g, g.ndata['x'])
                  loss = loss_fn(prediction, y)
                  train_loss += loss.item()
                  train_items += 1
                  train_iterator.set_postfix(loss=train_loss / train_items)
                  loss.backward()
                  optimizer.step()
                except:
                  pass
                  # print(mol)

        # Validation steps
        with torch.no_grad():
            model.eval()
            with tqdm.tqdm(validation_dataset,
                           desc=f'Validation Epoch {epoch}') as val_iterator:
                for mol, y in val_iterator:
                  try:
                      x = featurize_molecule(rdkit.Chem.MolFromSmiles(mol))
                      g = dgl.graph((x[1][0], x[1][1]))
                      g.ndata['x'] = x[0]
                      y = torch.tensor([y])
                      prediction = model(g, g.ndata['x'])
                      loss = loss_fn(prediction, y)
                      val_loss += loss.item()
                      val_items += 1
                      val_iterator.set_postfix(loss=val_loss / val_items)
                      val_iterator.set_postfix(loss=loss.item())
                  except:
                    pass
        train_losses.append(train_loss / train_items)
        val_losses.append(val_loss / val_items)

    return train_losses, val_losses
```
